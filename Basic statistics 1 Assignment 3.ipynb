```python
# Import Required Libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns


```

```python
# Data set loaded from csv by mentioning the path 
df = pd.read_csv(r"D:\data science\Assignments\sales_data_with_discounts.csv")
print(" Dataset Loaded Successfully!")
print(" Shape of Dataset:", df.shape)
print("\n First 5 Rows:")
display(df.head())
```

```python
#  Basic Information about Dataset 
numerical_cols = df.select_dtypes(include=['int64', 'float64']).columns
categorical_cols = df.select_dtypes(include=['object']).columns
```

```python
# Identify Numerical and Categorical Columns
print("\n Numerical Columns:", list(numerical_cols))
print(" Categorical Columns:", list(categorical_cols))
```

```python
# Descriptive Statistics for Numerical Columns
print("\n Descriptive Statistics:")
display(df[numerical_cols].describe())
```

```python
# Descriptive analytics shows mean, median, mode, and standard deviation for all numerical variables.
print("\n  Mean of Numerical Columns:")
print(df[numerical_cols].mean())
print("\n Median of Numerical Columns:")
print(df[numerical_cols].median())

print("\n Mode of Numerical Columns:")
print(df[numerical_cols].mode().iloc[0])

print("\n Standard Deviation of Numerical Columns:")
print(df[numerical_cols].std())
```

```python
# Histograms reveal the distribution  many variables are skewed ( positive,Negative or symmetric), indicating a few large transactions.
print("\n Plotting Histograms for Numerical Columns...")
df[numerical_cols].hist(figsize=(12, 10), bins=20, color='skyblue', edgecolor='black')
plt.suptitle(" Histogram: Distribution of Numerical Columns", fontsize=16)
plt.show()
```

```python
# Box plots help detect outliers  especially in sales, profit, or discount values
print("\n Box Plots for Outlier Detection...")
plt.figure(figsize=(12, 6))
sns.boxplot(data=df[numerical_cols])
plt.title(" Boxplot of Numerical Columns")
plt.xticks(rotation=45)
plt.show()
```

```python
# Bar charts show category distributions  highlighting dominant regions, product categories, or segments.
print("\n Bar Charts for Categorical Columns...")
for col in categorical_cols:
    plt.figure(figsize=(8, 4))
    df[col].value_counts().plot(kind='bar', color='teal', edgecolor='black')
    plt.title(f" Bar Chart - {col} Distribution")
    plt.xlabel(col)
    plt.ylabel("Count")
    plt.xticks(rotation=30)
    plt.show()
```

```python
# Summary 
# Descriptive Statistics:
# We calculated essential statistical measures  mean, median, mode, and standard deviation  for numerical columns.
# These metrics helped us understand the central tendency (typical values) and spread (variability) of the data.
# For instance, the mean provided the average value, while standard deviation indicated how much the data varied around that mean.

# Data Visualization:
# Histograms revealed the distribution of numerical data, showing whether it was normal, skewed, or contained outliers.
# Box plots helped detect outliers and understand the interquartile range (IQR), which represents the middle 50% of the data.
# Bar charts visualized the distribution of categorical variables, allowing us to see which categories were most frequent or dominant.

# Conclusion:
#The descriptive analytics and preprocessing steps provided a strong foundational understanding of the sales and discounts dataset. By combining statistical measures with effective visualizations, we identified key trends, data distributions, and potential outliers. This initial analysis is a crucial step in preparing the dataset for advanced analytics, predictive modeling, or decision-making processes.
```
