```python
import numpy as np
import scipy.stats as stats
```

```python
data = np.array([1.13, 1.55, 1.43, 0.92, 1.25, 1.36, 1.32, 0.85, 1.07, 
                 1.48, 1.20, 1.33, 1.18, 1.22, 1.29])
```

```python
n = len(data)
mean = np.mean(data)
s = np.std(data, ddof=1)
```

```python
 #Confidence level
confidence = 0.99
alpha = 1 - confidence

# Degrees of freedom
df = n - 1

# t critical value for 99% CI
t_critical = stats.t.ppf(1 - alpha/2, df)

# Standard Error
SE_t = s / np.sqrt(n)

# Margin of Error
ME_t = t_critical * SE_t

# Confidence Interval
CI_t_lower = mean - ME_t
CI_t_upper = mean + ME_t
```

```python
print("a) 99% Confidence Interval using sample standard deviation (t-distribution):")
print(f"Mean = {mean:.3f}")
print(f"Sample Std Dev (s) = {s:.3f}")
print(f"t-critical = {t_critical:.3f}")
print(f"Confidence Interval = ({CI_t_lower:.3f}, {CI_t_upper:.3f})")
print("-" * 60)
```

```python
# Given population standard deviation
sigma = 0.2

# z critical value for 99% CI
z_critical = stats.norm.ppf(1 - alpha/2)

# Standard Error
SE_z = sigma / np.sqrt(n)

# Margin of Error
ME_z = z_critical * SE_z

# Confidence Interval
CI_z_lower = mean - ME_z
CI_z_upper = mean + ME_z
```

```python
print("b) 99% Confidence Interval using known population standard deviation (z-distribution):")
print(f"Mean = {mean:.3f}")
print(f"Population Std Dev (σ) = {sigma:.3f}")
print(f"z-critical = {z_critical:.3f}")
print(f"Confidence Interval = ({CI_z_lower:.3f}, {CI_z_upper:.3f})")
```

```python
import numpy as np
import scipy.stats as stats
```

```python

X_mean = 600        # mean number of units produced in a week
sigma_X = 25        # standard deviation of units
sample_mean = 3050  # sample mean weekly cost
n = 25              # sample size
alpha = 0.05        # significance level



# STEP 1: Calculate theoretical mean weekly cost (μ)
# -----------------------------
mu = 1000 + 5 * X_mean
print(f"Theoretical Mean Weekly Cost (μ) = Rs. {mu:.2f}")

# -----------------------------
# STEP 2: Calculate population standard deviation of weekly cost (σ)
# -----------------------------
# Since W = 1000 + 5X, the standard deviation of W = 5 * σ_X
sigma = 5 * sigma_X
print(f"Population Standard Deviation of Weekly Cost (σ) = Rs. {sigma:.2f}")

# -----------------------------
# STEP 3: Calculate the Z Test Statistic
# -----------------------------
z = (sample_mean - mu) / (sigma / np.sqrt(n))
print(f"Z Test Statistic = {z:.3f}")

# -----------------------------
# STEP 4: Determine the critical value and p-value
# -----------------------------
z_critical = stats.norm.ppf(1 - alpha)   # one-tailed test (right tail)
p_value = 1 - stats.norm.cdf(z)

print(f"Critical Value (z_critical) = {z_critical:.3f}")
print(f"P-value = {p_value:.5f}")

# -----------------------------
# STEP 5: Decision Rule
# -----------------------------
if z > z_critical:
    decision = "Reject the Null Hypothesis (H0)"
else:
    decision = "Fail to Reject the Null Hypothesis (H0)"

print(f"Decision: {decision}")

# STEP 6: Conclusion
# -----------------------------
if z > z_critical:
    print("\nConclusion:")
    print("There is strong evidence at the 5% significance level that the average weekly operating cost")
    print("is higher than the cost predicted by the model.")
else:
    print("\nConclusion:")
    print("There is insufficient evidence to conclude that the average weekly operating cost")
    print("is higher than the cost predicted by the model.")
```

```python


# Step-by-Step Process

# The theoretical cost model predicts weekly cost as
# (W = 1000 + 5X).

# Based on (\bar{X} = 600), the theoretical mean cost = Rs. 4000.

# The population standard deviation of weekly cost =
(5 \times \sigma_X = 125).

# The hypothesis test checks whether the true mean cost exceeds Rs. 4000.

# Hypotheses

# (H_0: \mu = 4000)

# (H_1: \mu > 4000)

# Conclusion
# Since the test statistic (z) does not exceed the critical value, we fail to reject (H_0).
# There is no significant evidence that operating costs have increased beyond the model’s prediction.








```
